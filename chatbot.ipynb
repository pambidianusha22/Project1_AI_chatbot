```python
# Dictionary storing chatbot responses
chatbot = {

    # Greeting responses
    "hello": "Hi!",
    "hi": "Hello!",
    "hey": "Hey there!",

    # General responses
    "how are you": "I am fine!",
    "what is your name": "I am AI Chatbot.",

    # Exit response
    "bye": "Goodbye!"
}

# Display welcome message
print("Welcome to AI Chatbot!")

# Run chatbot continuously
while True:

    # Take user input
    user_input = input("You: ").lower()

    # Get response from dictionary
    response = chatbot.get(user_input)

    # If response exists
    if response:

        # Display chatbot response
        print("Bot:", response)

        # Stop chatbot when user types bye
        if user_input == "bye":
            break

    # If input not found in dictionary
    else:
        print("Bot: I don't understand.")
```
